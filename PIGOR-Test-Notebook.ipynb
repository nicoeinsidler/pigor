# PIGOR Test Notebook

In this notebook you'll find a few quick examples how to work with the Measurement class used by PIGOR in a Jupyter Notebook. Let's start by importing the Measurement module.

## Importing the Measurement Class

```python
import measurement
```

```python
import pathlib

home = pathlib.Path('.')
files = home.rglob('*.dat')

for f in files:
    print(f)
```

Now we have a list of available testfiles. Let's pick one and create a Measurement instance with it.

## Creating our first Measurement Instance

```python
m1 = measurement.Measurement(
    pathlib.Path('./testfiles/polarimeter/2018-11-23-1545-scan-dc2x.dat')
)
```

Note here that we have passed on a `pathlib.Path` object to `measurement.Measurement` instead of a string. This is because `measurement.Measurement` expects us to pass on a Path object.

We can now perform two main actions on `m1`:

- plot()
- contrast()

## Calculating the Contrast

```python
m1.contrast()
```

Calculating the contrast is straightforward. It simply returns a list of contrasts.

So let's plot it, too.

## Plotting a Measurement

```python
m1.plot(enable_jupyter=True)
```

Notice here that we have to set `enable_jupyter=True`. This will actually tell the Measurement class to fire the show() method in matplotlib. You should see the plot right in this notebook. If not, make sure that:

- `%matplotlib inline` is set in the first cell of the notebook
- to pass `enable_jupyter=True` on plot()

It is important to realize, that the Measurement class instance automatically fits the data when plotting. To disable that, you can set `fit=False`when using the plot method.

## Degree of Polarisation Measurement

To test a degree of polarisation measurement, we can either load a degree of polarisation measurement or force the Measurement class to take it as a degree of polarisation measurement by passing on `type_of_measurement` when creating a new instance. The values can be set to:

- DC
- POL
- POS

```python
m2 = measurement.Measurement(pathlib.Path('./testfiles/polarimeter/a/2018-11-23-1325-degree-of-pol.dat'))
```

```python
m2.plot(enable_jupyter=True)
```

```python
print(m2.pos_file_path)
m2.pos_data
```

## Changing the Fit Function

By passing on `type_of_fit` when creating a new instance, we can alter the fit function. Additionally it can be changed by changing the file name, or after it has been created by passing `fit_function` to the `fit()` method.

A list of all available fit function can be obtained by:

```python
m2.fit_function_list
```

We can see here, that only `'gauss'` has additional values, which are the fits bounds. The last best bound guess is always saved. To get the last fit function, that was used, simply type:

```python
m2.used_fit_function
```
